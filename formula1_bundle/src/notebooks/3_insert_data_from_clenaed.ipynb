```python
silver = spark.read.table('formula1.silver_001.f1_silver_vol_cleandataset')

try:
    gold_race = spark.read.table("formula1.gold_001.race")
    last_processed = gold_race.agg({"ingestion_timestamp": "max"}).collect()[0][0]
except Exception:
    last_processed = None

if last_processed:
    new_races = silver.filter(col("ingestion_timestamp") > last_processed)
else:
    new_races = silver

race_schema = StructType(fields=[
    StructField("raceId", IntegerType(), False),
    StructField("year", IntegerType(), True),
    StructField("round", IntegerType(), True),
    StructField("race_name", StringType(), True),
    StructField("race_date", DateType(), True),
    StructField("race_time", StringType(), True),
    StructField("race_url", StringType(), True)
])

df_race = new_races.select(
    "raceId", "year", "round", "race_name", "race_date", "race_time", "race_url", col("timestamp").alias("ingestion_timestamp")
).distinct()


df_race.write.format("delta").mode("append") \
     .option("mergeSchema", "true") \
     .option("delta.enableChangeDataFeed", "true") \
     .saveAsTable("formula1.gold_001.race")
```
